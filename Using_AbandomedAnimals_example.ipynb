```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

import matplotlib.font_manager as fm
import matplotlib
import matplotlib.pyplot as plt
font_path = 'C:\\windows\\Fonts\\malgun.ttf'
font_prop = fm.FontProperties(fname=font_path).get_name()
print(font_prop)
matplotlib.rc('font', family=font_prop)
# 날짜 변환


# 데이터 로드
try:
    df = pd.read_json('data20250531_20250730.json')
except FileNotFoundError:
    print("Error: data.json 파일을 찾을 수 없습니다. 파일 경로를 확인해주세요.")
    exit()

# 'happenDt'를 날짜 형식으로 변환
df['happenDt'] = pd.to_datetime(df['happenDt'], format='%Y%m%d')

print("데이터 로드 및 'happenDt' 전처리 완료.")
print("데이터 미리보기:")
print(df.head())
print(f"총 데이터 건수: {len(df)}")

# 1. 날짜별 총 유기 동물 발생 건수 추이
# 가장 기본적인 시계열 분석입니다.
print("\n1. 날짜별 총 유기 동물 발생 건수 추이")
daily_counts = df.groupby('happenDt').size().reset_index(name='count')

plt.figure(figsize=(15, 6))
sns.lineplot(x='happenDt', y='count', data=daily_counts)
plt.title('Daily Abandoned Animal Occurrences Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Occurrences')
plt.grid(True)
plt.show()

# 2. 'processState' (상태)별 유기 동물 발생 건수
# 특정 상태의 동물이 얼마나 되는지 확인
print("\n2. 'processState' (상태)별 유기 동물 발생 건수")
plt.figure(figsize=(12, 6))
sns.countplot(y='processState', data=df, order=df['processState'].value_counts().index, palette='viridis')
plt.title('Number of Abandoned Animals by Process State')
plt.xlabel('Number of Occurrences')
plt.ylabel('Process State')
plt.grid(axis='x')
plt.show()

# 3. 'kindFullNm' (품종/종류)별 유기 동물 발생 건수
# 어떤 품종/종류의 동물이 가장 많이 유기되는지 확인
print("\n3. 'kindFullNm' (품종/종류)별 유기 동물 발생 건수 (상위 10개)")
top_kinds = df['kindFullNm'].value_counts().head(10).index
plt.figure(figsize=(12, 7))
sns.countplot(y='kindFullNm', data=df, order=top_kinds, palette='magma')
plt.title('Top 10 Abandoned Animal Kinds')
plt.xlabel('Number of Occurrences')
plt.ylabel('Kind Full Name')
plt.grid(axis='x')
plt.show()

# 4. 'sexCd' (성별)별 유기 동물 발생 건수
print("\n4. 'sexCd' (성별)별 유기 동물 발생 건수")
plt.figure(figsize=(8, 5))
sns.countplot(x='sexCd', data=df, palette='cividis')
plt.title('Number of Abandoned Animals by Sex')
plt.xlabel('Sex Code')
plt.ylabel('Number of Occurrences')
plt.grid(axis='y')
plt.show()

# 5. 'neuterYn' (중성화 여부)별 유기 동물 발생 건수
print("\n5. 'neuterYn' (중성화 여부)별 유기 동물 발생 건수")
plt.figure(figsize=(8, 5))
sns.countplot(x='neuterYn', data=df, palette='plasma')
plt.title('Number of Abandoned Animals by Neutered Status')
plt.xlabel('Neutered (Y/N)')
plt.ylabel('Number of Occurrences')
plt.grid(axis='y')
plt.show()

# 6. 월별/요일별 유기 동물 발생 추이 (날짜에서 파생)
# 계절성이나 요일별 패턴이 있는지 확인
print("\n6. 월별 유기 동물 발생 추이")
df['month'] = df['happenDt'].dt.month
monthly_counts = df.groupby('month').size().reset_index(name='count')
plt.figure(figsize=(10, 6))
sns.barplot(x='month', y='count', data=monthly_counts, palette='GnBu')
plt.title('Monthly Abandoned Animal Occurrences')
plt.xlabel('Month')
plt.ylabel('Number of Occurrences')
plt.xticks(ticks=np.arange(0,12), labels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.grid(axis='y')
plt.show()

print("\n7. 요일별 유기 동물 발생 추이")
df['day_of_week'] = df['happenDt'].dt.day_name() # 요일 이름
# 요일 순서 지정
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_counts_by_week = df.groupby('day_of_week').size().reindex(day_order).reset_index(name='count')

plt.figure(figsize=(10, 6))
sns.barplot(x='day_of_week', y='count', data=daily_counts_by_week, palette='PuRd')
plt.title('Day of Week Abandoned Animal Occurrences')
plt.xlabel('Day of Week')
plt.ylabel('Number of Occurrences')
plt.grid(axis='y')
plt.show()

# 8. 지역별 (orgNm) 월별 발생 건수 (히트맵)
# 특정 지역의 특정 월에 발생이 많은지 패턴 확인
print("\n8. 지역별 (orgNm) 월별 발생 건수 히트맵 (상위 10개 지역)")
# 상위 10개 지역 선정
top_orgnms = df['orgNm'].value_counts().head(10).index
df_top_orgnms = df[df['orgNm'].isin(top_orgnms)].copy()

org_month_counts = df_top_orgnms.groupby(['orgNm', 'month']).size().unstack(fill_value=0)
# 월 순서가 숫자로 되어 있어 정렬이 잘 되지만, 원하는대로 '1월', '2월' 등으로 표시하고 싶다면
# columns = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'] 형태로 변경 후 reindex 가능
plt.figure(figsize=(14, 8))
sns.heatmap(org_month_counts, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
plt.title('Monthly Abandoned Animal Occurrences by Top 10 OrgNm')
plt.xlabel('Month')
plt.ylabel('Organization Name')
plt.show()

# 9. 지역별 (orgNm) 성별 발생 비율 (스택 바 차트)
print("\n9. 지역별 (orgNm) 성별 발생 비율 (상위 10개 지역)")
# 상위 10개 지역 선정
org_sex_counts = df_top_orgnms.groupby(['orgNm', 'sexCd']).size().unstack(fill_value=0)
org_sex_proportions = org_sex_counts.div(org_sex_counts.sum(axis=1), axis=0) # 비율 계산

org_sex_proportions.plot(kind='bar', stacked=True, figsize=(14, 7), cmap='Set2')
plt.title('Proportion of Abandoned Animals by Sex in Top 10 OrgNms')
plt.xlabel('Organization Name')
plt.ylabel('Proportion')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sex Code')
plt.tight_layout()
plt.show()
```
