```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json





# JSON 데이터를 DataFrame으로 변환
df = pd.read_json('predict_3_year_data.json')


# 'org_name'을 기준으로 내림차순 정렬 (확률이 높은 순서로 막대 그래프를 그리기 위함)
df_sorted = df.sort_values(by='predicted_probability_percent', ascending=False)


#한글 폰트 설정 (Windows 사용자용)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False



# 막대 그래프 그리기 (Seaborn)
plt.figure(figsize=(10, 6))
sns.barplot(x='org_name', y='predicted_probability_percent', data=df_sorted, palette='viridis')
plt.xlabel('기관명', fontsize=12)
plt.ylabel('예측 확률 (%)', fontsize=12)
plt.title('8월 유기동물 발생 예측 확률 top 5 지역', fontsize=14)
plt.ylim(90, 100) # y축 범위를 조정하여 차이를 더 명확하게 볼 수 있음
plt.xticks(rotation=45, ha='right') # x축 레이블 회전
plt.tight_layout() # 레이아웃 자동 조정
plt.show()

# 파이 차트 그리기 (Matplotlib)
# 파이 차트는 전체 비율을 볼 때 유용하지만, 여기서는 각 항목의 값이 너무 비슷하여 시각적으로 큰 차이를 보이지 않을 수 있습니다.
# 각 항목의 라벨과 해당 퍼센티지를 포맷팅합니다.
labels = [f'{row["org_name"]} ({row["predicted_probability_percent"]:.2f}%)' for index, row in df_sorted.iterrows()]
sizes = df_sorted['predicted_probability_percent']
```
